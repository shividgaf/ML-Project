```python
import pandas as pd
df = pd.read_csv('titanic_data_set.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import Imputer
```

```python
i = Imputer()
```

```python
df.iloc[:,5:6] = i.fit_transform(df.iloc[:,5:6])
```

```python
df.Embarked.value_counts()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
```

```python
df.iloc[:,4:5] = l.fit_transform(df.iloc[:,4:5])
```

```python
df.iloc[:,11:] = l.fit_transform(df.iloc[:,11:])
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=['Embarked','Pclass'])
```

```python
df.head()
```

```python
df['FamilySize']  = df['SibSp'] + df['Parch'] + 1
```

```python
df.head()
```

```python
irrelevant =  ['PassengerId','Name','SibSp','Parch','Ticket','Cabin']
df = df.drop(irrelevant,axis=1)
```

```python
df.head()
```

```python
df.to_csv('Titanic_prepared_data')
```
